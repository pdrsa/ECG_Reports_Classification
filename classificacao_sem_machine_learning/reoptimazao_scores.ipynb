```python
scoresAntigo = open("../data/resultados/sem_siglas/resultado_sem_siglas_scores.csv", "r")
scoresNovo   = open('../data/resultados/sem_siglas/resultado_sem_siglas_scores_better.csv', 'w+')
i = 1
for line in scoresAntigo:
    if i != 1:
        a = line.split(",")
        b = [int(a[i]) for i in range(len(a)) if i % 3 == 1 or i == 0]
        scoresNovo.write(str(b)[1:-1])
        scoresNovo.write("\n")
        i+=1
        if i % 10000 == 0:
            print(i)
    else:
        scoresNovo.write(line)
        i = 2

scoresAntigo.close()
scoresNovo.close()
```
